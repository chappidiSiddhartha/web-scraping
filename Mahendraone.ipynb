```python
import time
import requests
import pandas as pd
#The Given Page is Infinitely Scrolled Page,so we need to use Selenium
from selenium import webdriver
browser = webdriver.Chrome()
from bs4 import BeautifulSoup
# Telling Selenium to get the URL.
browser.get("https://www.mahindrafirstchoice.com/used-cars/delhi")

# Selenium script to scroll to the bottom, wait 3 seconds for the next batch of data to load, then continue scrolling.  It will continue to do this until the page stops loading new data.
lenOfPage = browser.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
match=False
while(match==False):
  lastCount = lenOfPage
  time.sleep(3)
  lenOfPage = browser.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
  if lastCount==lenOfPage:
    match=True

# Now that the page is fully scrolled, grab the source code.
source_data = browser.page_source
headers = {}
headers['User-Agent'] = "Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/49.0.2623.110 Safari/537.3"
# Throw your source into BeautifulSoup and start parsing!
soup = BeautifulSoup(source_data, 'html.parser')

alls = []#To store All values
year1=[]
make_of_car=[]
model=[]
price1=[]
location1=[]
dist1=[]
d_or_p=[]
type1=[]
owner1=[]

for d in soup.findAll('div', attrs={'class':"col-md-4 float-left col-12 general-car"}):
    year = d.find('div', attrs={'class':'buyer_header'})
    div=year.text
    year1.append(year.text.split()[0])
    #print(year)
    t=div.split()[1]
        
    #print(type(div))
    #alls[:] = [reviews.lstrip('\n') for reviews in alls]
    #alls[:] = [reviews.rstrip('\n') for reviews in alls]
    t = t.replace('\n', '')
    t = t.replace('\r', '')
    make_of_car.append(t)#make_of_car list
    s=div.split()[2]
    s = s.replace('\n', '')
    s = s.replace('\r', '')
    model.append(s)# model list
    price = d.find('span', attrs={'class':'car_price'})
    price1.append(price.text)# price list
    location=d.find('span', attrs={'class':'buyer_locations icon_set stock_location b'})
    location1.append(location.text)# location list
    
    owner=d.find("ul",attrs={"class":"spec-info-list"})
    #owner1.append(owner1[0].split())
    dist1.append(owner.text.split()[0]+owner.text.split()[1])#Distance List
    d_or_p.append(owner.text.split()[2])#Diesel or Petrol List
    type1.append(owner.text.split()[3])#Type of Model
    owner1.append(owner.text.split()[4]+owner.text.split()[5])#Owner

alls.append(year1)
alls.append(make_of_car)
alls.append(model)
alls.append(price1)
alls.append(location1)
alls.append(dist1)
alls.append(d_or_p)
alls.append(type1)
alls.append(owner1)    
print(len(year))
print(len(make_of_car))
print(len(model))
print(len(price1))
print(len(location1))
print(len(dist1))
print(len(d_or_p))
print(len(type1))
print(len(owner1))
print(alls)
#flatten = lambda l: [item for sublist in l for item in sublist]
df1=pd.DataFrame(year1)
df2=pd.DataFrame(make_of_car)
df3=pd.DataFrame(model)
df4=pd.DataFrame(price1)
df5=pd.DataFrame(location1)
df6=pd.DataFrame(dist1)
df7=pd.DataFrame(d_or_p)
df8=pd.DataFrame(type1)
df9=pd.DataFrame(owner1)
df=pd.concat([df1, df2, df3,df4,df5,df6,df7,df8,df9],axis=1)
#df.to_csv('mahendra.csv', index=False)
df.columns =['year', 'make_of_car', 'model', 'Price(INR)','location','Distance','Dieselorpetrol','Type','owner']
df.to_csv('mahendra.csv', index=False)
df
```
